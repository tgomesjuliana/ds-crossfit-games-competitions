# Imports

```python
import pandas as pd
import requests
import nbimporter
from nbimporter import NotebookLoader

loader = NotebookLoader('.')
module = loader.load_module('data_scraping_functions')

from data_scraping_functions import scraping_data_open_and_games
```

# Definitions

```python
years = range(2007, 2024)
competition = "games"
base_path = "C:/Users/Zi/Documents/data-science-crossfit/"
```

# Data Scraping

```python
dataframes_list = []

# Iterate over each year
for year in years:
    # Scrape data for the 'games' competition and the current year
    df_games_year = scraping_data_open_and_games(year, competition)
    # Add a new column 'genderId' with the value of 'divisionId' for consistency
    df_games_year['genderId'] = df_games_year['divisionId']
    # Drop irrelevant columns
    df_games_year = df_games_year.drop(columns=['postCompStatus', 'profilePicS3key', 'countryShortCode', 'teamCaptain', 'divisionId'])
    # Add a new column 'year' with the current year
    df_games_year['year'] = year
    # Append the data for the current year to the combined list
    dataframes_list.append(df_games_year)
# Create a single DataFrame by concatenating the list of DataFrames
df_games_conso = pd.concat(dataframes_list)
# Print the consolidated DataFrame
df_games_conso
```

# Exports

```python
df_games_conso.to_csv(base_path + "csv_files/version_1_raw_data/df_games_conso.csv", index=False)
```
